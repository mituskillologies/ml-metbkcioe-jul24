```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
df.shape
```

```python
# separate input and output

x = df[['hp', 'wt', 'gear']]
y = df['mpg']
```

```python
x
```

### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
sca = MinMaxScaler()
x_scaled = sca.fit_transform(x)
```

```python
x_scaled;
```

### Build the model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(x_scaled, y)
```

```python
lr.score(x_scaled, y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
hp = 130
wt = 4.1
gear = 5
```

```python
lr.predict(sca.transform([[hp, wt, gear]]))
```
