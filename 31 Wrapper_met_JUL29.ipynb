```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
df.head()
```

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x, y)
```

```python
print('accuracy:', lr.score(x, y))
```

### Recursive feature elimination

```python
from sklearn.feature_selection import RFE
```

```python
rfe = RFE(lr)
```

```python
x = df.drop('mpg', axis = 1)
```

```python
rfe.fit(x, y)
```

```python
rfe.get_support()
```

```python
x = x.iloc[:, rfe.get_support()]
```

```python
x
```

```python
lr = LinearRegression()
lr.fit(x, y)
print('accuracy:', lr.score(x, y))
```

# backward/forward

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
sfs = SequentialFeatureSelector(lr, direction= 'backward')
```

```python
x = df.drop('mpg', axis = 1)
```

```python
sfs.fit(x, y)
```

```python
x = x.iloc[:, sfs.get_support()]
```

```python
x
```

```python
lr = LinearRegression()
lr.fit(x, y)
print('accuracy:', lr.score(x, y))
```
