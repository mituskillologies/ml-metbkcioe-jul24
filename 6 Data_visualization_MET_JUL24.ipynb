### Bar plot

```python
students = [56, 60, 73, 40, 80]

courses = ['java', 'python', 'C++', 'kotlin', 'C']
```

```python
colors = ['cyan', 'orange', 'purple', 'yellow', 'skyblue']
```

```python
plt.title('Student Enrollment')
plt.xlabel('Courses')
plt.ylabel('Students')
bar = plt.bar(courses, students, color = colors, edgecolor = 'black', width = 0.6,
       linewidth = 3, alpha = 0.9);
plt.bar_label(bar)
plt.grid();
```

```python
df1 = pd.read_csv('Social_Network_Ads.csv')
```

```python
'''
1. line plot (student.csv)
2. bar plot (student.csv => student name, marks)
3. scatter plot (Social_Network_Ads.csv => age, estimated salary)
4. pie chart (Social_Network_Ads.csv => how many purchased/not purchased the product)
'''
```

```python
df1
```

```python
df1['Purchased'].value_counts()
```

```python
plt.style.use('default')
plt.figure(figsize=(16, 9))

plt.subplot(2, 2, 1)
plt.title('Line Plot')
plt.xlabel('roll no.')
plt.ylabel('marks')
plt.xticks(range(0, 11, 1))
plt.plot(df['roll'], df['marks'], marker = 'o', mfc = 'red', ls = ':', ms = 5,
        color = 'black');
plt.grid()

plt.subplot(2, 2, 2)
plt.title('bar plot')
plt.xlabel('student name')
plt.ylabel('marks')
bar = plt.bar(df['name'], df['marks'],edgecolor = 'black', width = 0.6,
       linewidth = 3, alpha = 0.9);
plt.bar_label(bar)
plt.grid();

plt.subplot(2, 2, 3)
plt.scatter(df1['EstimatedSalary'], df1['Age'], c = df1['Purchased'])
plt.grid()

plt.subplot(2, 2, 4)
plt.pie(df1['Purchased'].value_counts(), labels = df1['Purchased'].value_counts().index,
       autopct = '%.2f%%', 
       labeldistance= 1.1, startangle= 90, shadow = True,
        radius = 1.3);
```

```python
q3 = np.percentile(df1['Age'], 75)
q1 = np.percentile(df1['Age'], 25)
```

```python
q3, q1
```

```python
iqr = q3 - q1
```

```python
iqr
```

```python
ub = q3 + (1.5 * iqr)

lb = q1 - (1.5 * iqr)

ub, lb
```
