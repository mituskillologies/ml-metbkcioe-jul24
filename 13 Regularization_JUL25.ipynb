```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BostonHousing.csv')
df.head()
```

```python
x = df.drop('medv', axis = 1)
y = df['medv']
```

```python
x
```

### feature scaling

```python
from sklearn.preprocessing import StandardScaler
sca = StandardScaler()
x_scaled = sca.fit_transform(x)
```

### cross validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_scaled, y,
                                                   random_state= 0,
                                                   test_size= 0.2)
```

### build linear regression model

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)  # testing accuracy
```

```python
lr.score(x_train, y_train)   # training accuracy
```

### polynomial reg

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
x_poly = poly.fit_transform(x_scaled)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x_poly, y,
                                                   random_state= 0,
                                                   test_size= 0.2)
```

```python
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
print('training acc:', lr.score(x_train, y_train))
print('testing acc:', r2_score(y_test, y_pred))
```

### Ridge Regression

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha = 18)
```

```python
ridge.fit(x_train, y_train)
```

```python
print('training acc:', ridge.score(x_train, y_train))
print('testing acc:', ridge.score(x_test, y_test))
```

### Lasso reg

```python
from sklearn.linear_model import Lasso
las = Lasso(alpha = 0.2)
```

```python
las.fit(x_train, y_train)
```

```python
print('training acc:', las.score(x_train, y_train))
print('testing acc:', las.score(x_test, y_test))
```
