```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('movies.csv')
df.head()
```

```python
x = df[['Budget', 'Celebrities']]
y = df['Class']
```

##### visualize the data

```python
sns.jointplot(x = 'Celebrities', y = 'Budget', data = df,
             hue = 'Class');
```

#### build the model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=0)
```

```python
dt.fit(x, y)
```

```python
# predict

new = [[3000000, 10]]
```

```python
dt.predict(new)
```
